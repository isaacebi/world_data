```python
URL = baseURL + locationKK + "1"

r = requests.get(URL) 

soup = BeautifulSoup(r.text, 'html.parser')



# Find all <div> elements with class "center-block"
target_ = soup.find_all('a')

news = []
for item in target_:
    attribute = item.get('href')
    if attribute and "/news" in attribute:
        news.append(attribute)
```

```python
contents = soup.find_all('div', class_='content')

for content in contents:
    title = content.find('div', class_='title')
    
    date = content.find('div', class_='newstag')

    news = {
        'date': date.text.split('\n')[1].strip(),
        'location': date.text.split('\n')[2].strip(),
        'title': title.text.strip()
    }
    print(news)
```

```python
URL = "https://www.dailyexpress.com.my/news/226598/we-haven-t-showered-ums-students-greet-zambry-with-protest-banners-over-water-issues/"
r = requests.get(URL) 
soup = BeautifulSoup(r.text, 'html.parser')
target_ = soup.find_all('span', class_="newsContent")

texts = []

for item in target_:
    texts.append(item.text)
    print(item.text)
```

```python
import re

# Remove "ADVERTISEMENT"
ad_pattern = re.compile(r'ADVERTISEMENT')

# Remove SPONSORED CONTENT section and everything after it
content_pattern = re.compile(r'Stay up-to-date.*$', re.MULTILINE | re.DOTALL)

# Apply the patterns
text = ad_pattern.sub('', texts[0])
text = content_pattern.sub('', text)


print(text)
```
